# 🛠️ Patch Existing Probe Features

This notebook adds the `has_probe_data` flag to your existing `feat_probe.parquet` file without requiring you to re-run the entire feature extraction script.

```python
import pandas as pd
import os

# Path to your current feature file
FILE_PATH = '../../data/processed/features/feat_probe.parquet'

if not os.path.exists(FILE_PATH):
    print(f"❌ File not found at {FILE_PATH}")
else:
    # 1. Load data
    df = pd.read_parquet(FILE_PATH)
    print(f"✅ Loaded {len(df):,} segments.")

    # 2. Add the flag
    # 1 if there was at least one ping across all time bins, 0 otherwise
    df['has_probe_data'] = (df['probe_count'] > 0).astype(int)
    
    print(f"📊 Probe Data Coverage: {df['has_probe_data'].mean()*100:.1f}%")
    print(f"📊 Count: {df['has_probe_data'].value_counts().to_dict()}")

    # 3. Save back to the same location
    df.to_parquet(FILE_PATH, index=False)
    print(f"🚀 Successfully updated {FILE_PATH}")
```

### Verification
Run this to make sure the column is there now.

```python
test_df = pd.read_parquet(FILE_PATH)
print("New Columns:", [c for c in test_df.columns if 'probe' in c])
test_df[['segment_id', 'probe_count', 'has_probe_data']].head(10)
```
